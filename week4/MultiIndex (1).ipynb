## Create A MultiIndex with the `.set_index()` Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"])
bigmac.head(3)
```

```python
bigmac.set_index(keys = ["Date", "Country"], inplace = True) #indicando que essas duas colunas serão índice
bigmac.head(3)
```

```python
bigmac.sort_index(inplace = True) #ordenar o índice
```

```python
bigmac.head(3)
```

```python
bigmac.index.names #nomes das colunas que são índices
```

```python
type(bigmac.index) #tipo de objeto que é a coluna index
```

```python
bigmac.index[0] #valores dos index para a primeira linha
```

## The `.get_level_values()` Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.index.get_level_values(0) #traz os dados da coluna 0 do index
```

```python
bigmac.index.get_level_values("Date") #essa notação é equivalente à anterior
```

```python
#bigmac.index.get_level_values(1)
bigmac.index.get_level_values("Country")
```

## The `.set_names()` Method on MultiIndex

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

## Extract Rows from a `MultiIndex DataFrame`

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.loc[("2010-01-01", "Brazil"), "Price in US Dollars"] #para acessar a informação, é necessário informar os valores dos
                                                            #dois índices
```

```python
bigmac.loc[("2015-07-01", "Chile"), "Price in US Dollars"]
```

```python
bigmac.ix[("2016-01-01", "China"), 0]
```

## The `.transpose()` Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.transpose() #para transpor a tabela 
bigmac.head(1)
```

```python
bigmac.loc["Price in US Dollars", ("2016-01-01", "Denmark")] #para acessar a informação com multi index como linhas
```

## The `.swaplevel()` Method

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.swaplevel() #mesclar e 'desmesclar' a célula que contém a mesma informação em várias linhas
bigmac.head(3)
```

## The `.sort_index()` Method on a MultiIndex `DataFrame`

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.sort_index(ascending = [True, False], inplace = True) #False no segundo elemento: colocar em ordem decrescente a 
                                                             #segunda coluna de index
```

```python
bigmac.head(3)
```

## The `pivot` Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales.pivot(index="Date", columns="Salesman", values="Revenue").head(3)
```

## The `.stack()` Method

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
world.stack().to_frame() #uma outra forma de apresentar os dados na tabela
```

## The `.unstack()` Method, Part 3

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
s = world.stack()
s.head(3)
```

```python
s.unstack(level = ["year", "country"])
```

```python
s = s.unstack("year", fill_value = 0)
```

```python
s.head()
```
