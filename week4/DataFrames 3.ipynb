```python
bond.where(bond["Box Office"] > 800)
```

```python
mask2 = bond["Box Office"] > 800
```

```python
bond.where(mask & mask2)
```

## A Review of the `.apply()` Method on Single Columns

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number): #definir uma função
    return str(number) + " MILLIONS!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions) #usar o método apply para aplicar essa função em toda
                                                                    #a coluna
```

```python
bond.head(3)
```

## The `.apply()` Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):             #definir uma função:
    
    actor = row[1] #actor corresponde à coluna 1 
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis = "columns")
```
