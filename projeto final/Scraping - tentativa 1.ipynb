```python
#Importando as bibliotecas necessárias 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Dessa url que fez-se o webscrapping

url = 'https://www.astro.com/astro-databank/Roberts%2C_Julia' 
```

```python
page = requests.get(url) #solicitando GET ao servidor
```

```python
#analisando a página com a biblioteca beautiful soup
soup = BeautifulSoup(page.content, 'html.parser') 
```

```python
# identificando a tag e fazendo a separação dos dados
rows = soup.find_all('tr') 
rows = [row.text.strip().split("\n") for row in rows]
```

```python
rows
```

```python
# combinando os dados em um dataframe utilizando a biblioteca pandas
data = rows

df = pd.DataFrame(data)
df.head()
```
