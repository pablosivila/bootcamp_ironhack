### Webscraping:

```python
#Importando as bibliotecas necessárias 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Dessa url que fez-se o webscrapping
# O objetivo é obter uma tabela com os lançamentos de filme do segundo semestre
url = 'https://www.guiadasemana.com.br/cinema/noticia/lancamentos-de-filmes-em-2019' 
```

```python
page = requests.get(url) #solicitando GET ao servidor
```

```python
#analisando a página com a biblioteca beautiful soup
soup = BeautifulSoup(page.content, 'html.parser') 
```

```python
# identificando a tag e fazendo a separação dos dados
rows = soup.find_all('p') 
rows = [row.text.strip().split("\r") for row in rows]
```

```python
# combinando os dados em um dataframe utilizando a biblioteca pandas
data = rows

df = pd.DataFrame(data)
df.head()
```

```python
# As linhas 0 a 2 correspondiam a dados que não eram da tabela de interesse. 
# Sendo assim, fez-se a exclusão dessas linhas, mantendo apenas os dados de interesse.
for i in range (2):
    df.drop([i], axis=0, inplace = True)
```

```python
# resetando o index e tirando a coluna "index"
df.reset_index(inplace=True)
df.drop(['index'], axis=1, inplace = True)
```

```python
# Fez-se uma classe, para acessar todos os filmes lançados a partir de 2014 até 2020:
```

```python
class filmes:
    """
    Essa é uma classe que acessa as informações sobre os títulos lançados de 2014 a 2020.
    """
    def __init__(self, url_pattern, sleep_interval=-1, content_parser=None):
        self.url_pattern = url_pattern
        self.sleep_interval = sleep_interval
        self.content_parser = content_parser
        self.lst=[] #criando uma lista vazia
   
    def scrape_url(self, url):
    
        response = requests.get(url)
        result = self.content_parser(response.content)
        self.lst.append(self.output_results(result)) #o output vai ser agregado na lista vazia criada anteriormente
   
    def output_results(self, r):
        return r 
       
    def kickstart(self):
        for i in range(14, 21): #para ler todas as as urls, usando a função scrape_url, no range de 2014-20
            self.scrape_url(self.url_pattern % i)
    
    def retorno(self):
        return self.lst #retorne a lista

URL_PATTERN = 'https://www.guiadasemana.com.br/cinema/noticia/lancamentos-de-filmes-em-20%s'
# regex pattern - substituir de 14 a 20 no %s


def quotes_parser(content): #limpando os dados obtidos
    soup = BeautifulSoup(content,'html.parser')
    rows =soup.find_all('p')
    for elem in soup.find_all(["a", "p", "div", "h3", "br"]):
        elem.replace_with(elem.text + "\n\n")
    rows = [row.text.strip().split("\n") for row in rows]
    
    return rows
 

my_spider = filmes(URL_PATTERN, content_parser=quotes_parser)


# Faz o scrapping
my_spider.kickstart()

#A variável "listas" salva uma lista de listas: cada uma dessas listas corresponde a um ano
listas=my_spider.retorno()
```
