```python
janken = ["Stone", "Paper", "Scissors"]

import random

def choices():
    while True:
        a=input("Please, enter your choice: A) Stone; B) Paper; C) Scissors")
        if a == "A" or a == "B" or a == "C":
            break
    return a
        
```

```python
def combat(a, b): #a: human choice; b: machine random choice
    
    if b=="Stone" and a == "A":
        saida= "0"

    elif b=="Stone" and a == "B":
        saida= "2"

    elif b=="Stone" and a == "C":
        saida="1"
    
    elif b=="Paper" and a == "B":
        saida= "0"

    elif b=="Paper" and a == "A":
        saida= "1"
    
    elif b=="Paper" and a == "C":
        saida= "2"

    elif b=="Scissors" and a == "C":
        saida= "0"

    elif b=="Scissors" and a == "A":
        saida ="2"

    elif b=="Scissors" and a == "B":
        saida= "1"
              
    return saida


```

```python
st=0 #game status
games = 5 #maximum number of games
win = 3 #number of games a player must win to win

def ch2 (st, games):
    while st <games:
        st=st+1
        return st
```

```python
mw=0 #machine wins
hw=0 #human wins
tie=0

while True:
    machine=random.choice(janken)
    c=choices()  
    d=combat(c,machine) #combat() result
    st=ch2(st,games)
          
    print(d)
    print ("Machine:",machine, "|  Player:",c)
    print("Status: game",st)
    
    if d=="0":
        tie=tie+1
        print("It's a tie! Ties:", tie)
    elif d=="1":
        mw=mw+1   
    elif d=="2":
        hw=hw+1
        
    print("Machine wins:", mw, "  x  Human wins:", hw)
    
         
    if hw==win:
        break
    elif mw==win:
        break
    elif st==games:
        break
    

if mw>hw:
    print("Winner: Machine")
elif hw>mw:
    print("Winner: Human")
else:
    print("Tie")
```

```python
def win2(g): #number of games a player must win to win
    if g==1:
        return 1
    elif g>1:
        return g-1

```

```python
st2=0 #game status

def ch2 (st2, g):
    while st2 <g:
        st2=st2+1
        return st2
```
